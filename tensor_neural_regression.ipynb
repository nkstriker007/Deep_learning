```python
#steps in modelling with tensor flow
# 1) creating a model -define the input,output layer as well as hidden layes of a deep learning model
# 2) compiling a model-define the loss function(function tells how wrong it is) and the optimizer  (tells our model how to improve the patterns it is learning and ecvaluation metrics (what we can use to interpret the performance of the model))
# 3) Fitting a model -letting the model find the relationship between the input and output values
X=tf.cast(tf.constant(X),dtype=tf.float32)
y=tf.cast(tf.constant(y),dtype=tf.float32)
    #only one layer because for one x we need to get only on y

#epoches -no of oppurtunity
tf.random.set_seed(42)

# Create a model using the Sequential API
model = tf.keras.Sequential([
  tf.keras.layers.Dense(1),
  tf.keras.layers.Dense(1).
])

# Compile the model
model.compile(loss=tf.keras.losses.mae, # mae is short for mean absolute error
              optimizer=tf.keras.optimizers.SGD(), # SGD is short for stochastic gradient descent
              metrics=["mae"])

# Fit the model
# model.fit(X, y, epochs=5) # this will break with TensorFlow 2.7.0+
model.fit(tf.expand_dims(X, axis=-1), y, epochs=5)
model.predict([17.0])
```

```python
def plot_predictions(train_data=xtrain_data,
                 train_labels=ytrain_data,
                 test_data=xtest_data,
                 test_labels=ytest_data,
                 predictions=y_pred):
  """
  Plots training data, test data and compares predictions.
  """
  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c="b", label="Training data")
  # Plot test data in green
  plt.scatter(test_data, test_labels, c="g", label="Testing data")
  # Plot the predictions in red (predictions were made on the test data)
  plt.scatter(test_data, predictions, c="r", label="Predictions")
  # Show the legend
  plt.legend();
plot_predictions(train_data=xtrain_data,
                 train_labels=ytrain_data,
                 test_data=xtest_data,
                 test_labels=ytest_data,
                 predictions=y_pred)
```

```python
#functionizing mae and mse
def mae(y_true=ytest_data,y_pred=y_pred):
  return tf.metrics.mean_absolute_error(y_true=y_true,y_pred=tf.squeeze(y_pred))
def mae(y_true=ytest_data,y_pred=y_pred):
  return tf.metrics.mean_squared_error(y_true=y_true,y_pred=tf.squeeze(y_pred))
```

```python
import matplotlib.pyplot as plt
import pandas as pd
insurance=pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance["smoker"],insurance["age"]
```

```python
objects=insurance.select_dtypes(include=[object])
print(objects)
insurance=pd.get_dummies(insurance)
insurance.head()
charges=insurance["charges"]
insurance=insurance.drop("charges",axis="columns")
```

```python
insurance
from sklearn.model_selection import train_test_split
x=insurance
y=charges
print(x,y)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
len(x_train),len(x_test),len(y_test)
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
insurance_model=tf.keras.Sequential(
    [tf.keras.layers.Dense(100,activation="relu"),
      tf.keras.layers.Dense(10,activation="relu"),

     tf.keras.layers.Dense(1)]
)
insurance_model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=["mae"])
history=insurance_model.fit(x_train,y_train,epochs=200,callbacks=[callback])
```

```python
#plot history loss curve
insurance_model.evaluate(x_test,y_test)
pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```

##preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
tf.random.set_seed(42)
df=pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
ct=make_column_transformer(
    (MinMaxScaler(),["age","bmi","children"]),
    (OneHotEncoder(handle_unknown="ignore"),["sex","smoker","region"])
)
x=df.drop("charges",axis=1)
y=df["charges"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
ct.fit(x_train)
x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
```

```python
x_train_normal.shape,x_train.shape
model=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(10,activation="relu"),
    tf.keras.layers.Dense(1)
])
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.Adam(),metrics=["mae"])
history=model.fit(x_train_normal,y_train,epochs=100)
model.evaluate(x_test_normal)
```

```python
model.evaluate(x_test_normal,y_test)
pd.DataFrame(history.history).plot()
```
